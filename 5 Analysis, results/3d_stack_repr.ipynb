```python
img = stack[10]
io.imshow(img)
img.shape
```

```python
T = np.array([[1,-1, 0],
              [0, 1, 0],
              [0, 0, 1]])
transf = transform.AffineTransform(matrix = T)
h, w, c = img.shape
pad = 100
bimg = np.zeros([h+pad*2, w+pad*2, c], dtype=img.dtype)
for i in range(3):
    bimg[:,:,i] = np.pad(img[:,:,i], pad_width = pad, mode='constant', constant_values=(0,0))
# bimg = skimage
io.imshow(transform.warp(bimg, transf))
```
