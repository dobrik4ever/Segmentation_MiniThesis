```python
stack = io.imread('../4 Raw data/good stacks/160719_06 stack_16-03-20/classified mask overlay.tif')
print(stack.shape)
z = 90
x0, y0 = 10, 85
s = 40
w = stack[z,:, y0:y0+s, x0:x0+s]
mask = w[0] / -255.0 + 1.0
img  = w[1]
plt.imshow(img, cmap='gray')
plt.imshow(mask, cmap='viridis', alpha=.5)
plt.colorbar()
```

```python
from skimage import measure
contours = measure.find_contours(mask, 0.5)
plt.imshow(mask)
for contour in contours:
    plt.plot(contour[:, 1], contour[:, 0], linewidth=2)
    
```
